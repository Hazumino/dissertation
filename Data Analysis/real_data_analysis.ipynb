```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
```

```python
first_lockdown = {
    'France':['2020-03-17','2020-05-11'],
    'Italy' :['2020-03-09','2020-05-18'], 
    'Spain' :['2020-03-14','2020-05-09'], 
    'Germany' :['2020-03-16','2020-05-30'], 
    'UK' :['2020-03-23','2020-07-04'], 
}
second_lockdown = {
    'France':['2020-10-30','2020-12-15'],
    'Italy' :['2020-12-24','2021-01-06'], 
    'Spain' :['2020-03-03','2020-03-03'], 
    'Germany' :['2020-11-02','2021-03-01'], 
    'UK' :['2020-11-05','2021-03-28'],  
}

first_border_closure = {
    'France':['2020-10-30','2020-12-15'],
    'Italy' :['2020-12-24','2021-01-06'], 
    'Spain' :['2020-03-03','2020-03-03'], 
    'Germany' :['2020-11-02','2021-03-01'], 
    'UK' :['2020-11-05','2021-03-28'],  
}
```

```python
country_names = ['France','Germany','Italy','Spain','UK']
nameOfFiles = "../processed_data/cumDeath"

for x in country_names:
    fLock = first_lockdown[x]
    sLock = second_lockdown[x]
    t = np.linspace(0, 600, 600)
    currName = nameOfFiles + x +'.csv'
    df = pd.read_csv(currName)
    deaths = df["cumDeath"].tolist()
    fig = plt.figure(facecolor='w')
    ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True, label="Deaths")
    ax.plot(df['date'], deaths)
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=130))
    plt.gcf().autofmt_xdate()
    plt.title(x + " Cumulative Covid-19 Deaths")
    plt.xlabel("Time / Days")
    plt.ylabel("Cumulative Deaths")
    plt.axvspan(fLock[0],fLock[1], color='red', alpha=0.5)
    plt.axvspan(sLock[0],sLock[1], color='green', alpha=0.5)
    ax.legend(['CumDeath','First Lockdown','Second Lockdown'])
    ax.grid(b=True, which='major', c='w', lw=2, ls='-')
```
