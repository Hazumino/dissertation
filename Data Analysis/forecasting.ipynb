```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
t = np.linspace(0, 1000, 1000)
ax.plot(t, np.linspace(0, 1000, 1000), 'b', alpha=0.5, lw=2, label='Deceased-Predicted')
ax.plot(t, np.linspace(500, 1000, 1000), 'r', alpha=0.5, lw=2, label='Deceased-Real',linestyle='dashed')
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
ax.set_xlabel('Time / days')
ax.set_ylabel('Accumulative Deceased')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
```
